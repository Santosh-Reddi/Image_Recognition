```python
#Preprocessing
import os #to move between folders
import matplotlib.pyplot as plt #to display image
import numpy as np #for numerical computing
from skimage.io import imread #to read images
from skimage.transform import resize #to resize images to same size

target=[] #output
images = [] #input
flat_data = [] 

DATADIR = '/content/images'
CATEGORIES = ['pretty sunflower','rugby ball leather','ice cream cone']


for category in CATEGORIES:
  class_num= CATEGORIES.index(category) #getting index number of values in CATEGORIES array
  path=os.path.join(DATADIR,category) #create directory path 
  for img in os.listdir(path):
    img_array=imread(os.path.join(path,img))
    #print(img_array.shape) #prints the shape of image (m,n,p) m=width,n=height,p=type of image{if 3 its RGB image}
    #plt.imshow(img_array) #show the actual image
    img_resized= resize(img_array,(150,150,3)) #skimage Normalizes the pixel values
    flat_data.append(img_resized.flatten())
    images.append(img_resized)
    target.append(class_num)

flat_data=np.array(flat_data)
target=np.array(target)
images=np.array(images)
```

```python
#Split data in Train and Test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(flat_data,target,test_size=0.3,random_state=109)
```

```python
#Parameter tuning using GridSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn import svm

param_grid=[{'C':[1,10,100,1000],'kernel':['linear']},
            {'C':[1,10,100,1000],'gamma':[0.001,0.0001],'kernel':['rbf']}
            ]

svc=svm.SVC(probability=True) #returns the probability of other categories as well
clf = GridSearchCV(svc,param_grid)
clf.fit(x_train,y_train)
```

```python
y_pred= clf.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```

```python
# Save the model using Pickle Library
import pickle
pickle.dump(clf,open('img_model.p','wb'))  #write the file
#file is save in main folder
```
